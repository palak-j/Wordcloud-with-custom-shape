```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
import pickle

from PIL import Image
```

```python
# Read input file. Here, it is in the pickle format

with open('vehicle.p', 'rb') as f:
    data = pickle.load(f)
```

```python
data
```

```python
# Read image as masked format (Wordcloud will be in the shape of this image)

mask = np.array(Image.open('car.jpg'))
```

```python
# Take data for wordcloud (here, we are taking all the manufacturers from the vehicle data)
text = ' '.join(data['manufacturer'].tolist())
```

```python
# Create wordcloud 
wordcloud = WordCloud(width = 7000, height = 5000, random_state=1, prefer_horizontal=1,background_color='white', contour_color='cornflowerblue', colormap='Set2', contour_width=4, collocations=True, mask=mask).generate(text)
```

```python
# parameters of the plot

plt.figure(figsize=(20, 10))
plt.imshow(wordcloud) 
plt.tight_layout(pad=0)
plt.axis("off");
```
